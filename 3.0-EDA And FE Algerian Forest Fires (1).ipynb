## Algerian Forest Fires Dataset

12. Classes: two classes, namely Fire and not Fire

The Fire Weather Index (FWI) is a numerical rating system used to estimate the risk of forest fires. It is part of the Canadian Forest Fire Weather Index System, which is widely adopted in various countries for wildfire management. The FWI system provides a measure of the potential fire intensity and spread rate based on weather conditions. Here's a brief overview of its components:

Fine Fuel Moisture Code (FFMC): This index measures the moisture content of fine surface fuels such as grass, leaves, and twigs. Lower values indicate wetter conditions, which reduce fire ignition probability, while higher values suggest drier conditions, increasing ignition risk.

Duff Moisture Code (DMC): This code estimates the moisture content in the loosely compacted layer of organic material found beneath the fine surface fuels. It affects the smoldering potential and the ease with which fires can spread.

Drought Code (DC): The DC measures the moisture content in deep, compact organic layers. It reflects long-term drying trends and is important for understanding the potential for sustained fire activity.

Initial Spread Index (ISI): This index predicts the expected rate of fire spread immediately after ignition, based on the combined effects of wind and the FFMC.

Buildup Index (BUI): The BUI is derived from the DMC and DC, representing the total amount of fuel available for combustion.

Fire Weather Index (FWI): The FWI is a composite index that provides a general indication of fire intensity. It integrates the ISI and BUI to offer a comprehensive assessment of fire danger, assisting in wildfire management decisions.

The FWI values are calculated daily using temperature, relative humidity, wind speed, and precipitation data. Higher FWI values signify greater fire risk and potential fire severity.

The Fire Weather Index (FWI) does not provide an absolute threshold that guarantees a fire will occur, as it is primarily a measure of potential fire intensity and spread rather than a direct predictor of fire ignition. However, higher FWI values indicate a greater likelihood and potential severity of fires if ignition sources are present.

The interpretation of FWI values typically falls into the following categories:

Low (0-5): Low fire risk. Fires are unlikely to start, and if they do, they will likely be of low intensity and easy to control.
Moderate (6-12): Moderate fire risk. Fires can start and spread, but they are generally controllable.
High (13-25): High fire risk. Fires can start easily and spread rapidly, becoming difficult to control.
Very High (26-49): Very high fire risk. Fires start easily, spread very quickly, and are very challenging to control.
Extreme (50+): Extreme fire risk. Fires start easily, spread rapidly, and are extremely difficult to control.

## Percentage for Pie Chart

```python
percentage=df1['Classes'].value_counts(normalize=True)*100
percentage
```

value_counts(normalize=True)

Purpose: This method counts the occurrences of each unique value in the 'Classes' column.
Effect:
Without the normalize=True argument, value_counts() would return the counts of each unique value.
With normalize=True, it returns the relative frequencies (proportions) of each unique value instead of the raw counts. The proportions are given as decimal values, where the sum of all proportions equals 1.
* 100

Purpose: This multiplies the relative frequencies by 100.
Effect: Converts the proportions (which are between 0 and 1) into percentages.

## Plotting piechart

```python
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(5,10))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%',colors=["purple", "orange"])
plt.title("Pie Chart of Classes")
plt.show()
```
